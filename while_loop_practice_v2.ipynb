Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
#write answer here
a=1
b=20
while a<=b:
  print(a)
  a+=1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
#write answer here
num=int(input("enter any number: "))
i=1
total=0
while i<=num:
  total=total+i
  i=i+1

print("sum=", total)
```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here

n=int(input("enter any number: "))
rev=0
while n>0:
  digit=n%10
  rev=rev*10+digit
  n=n//10
print (rev)
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here

n = int(input("Enter a number: "))

original = n
reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")

```

Q8. Guess the number game.

Approach:
- Use while loop
- Continue until correct guess

```python
#write answer here
num=16
guess_num=int(input("enter the number: "))
while guess_num!=num:
  if guess_num<num:
    print("To low")
    guess_num=int(input("enter the number: "))
  else:
    print("Too high")
    guess_num=int(input("enter the number: "))
print("correct guess")
```
